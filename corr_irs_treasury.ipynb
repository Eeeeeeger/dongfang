```python
treasury = w.edb("S0059744,S0059745,S0059746,M0057946,S0059747,M0057947,S0059748,M1000165,M1004678,S0059749", "2015-02-28", "2023-02-28","Fill=Previous", usedf=True)[1]
treasury.columns=[f'到期收益率:{ii}年' for ii in range(1,11)]
treasury.index = pd.to_datetime(treasury.index)
treasury = treasury.sort_index().diff().dropna()
```

```python
import scipy.stats as stats
res = pd.DataFrame(columns=['fr007','treasury','lead_lag','correlation','data_length'])
for each_treasury in treasury.columns:
    for each_fr007 in fr007.columns:
        v = pd.concat([fr007[each_fr007],treasury[each_treasury]], axis=1).dropna()
        correlation, pvalue = stats.spearmanr(v.iloc[:,0], v.iloc[:,1])
        if pvalue<0.01:
            res= res.append(pd.Series({'fr007':each_fr007, 'treasury':each_treasury,"lead_lag":0,
                                       "correlation": correlation, 'data_length':len(v)}),
                      ignore_index=True)
res
```

```python
res.loc[res['correlation'].astype(float).abs().nlargest(5).index]
```

```python
params = list(range(-10,0)) + list(range(1,10))
for lead_lag in params:
    for each_treasury in treasury.columns:
        for each_fr007 in fr007.columns:
            v = pd.concat([fr007[each_fr007].shift(lead_lag),treasury[each_treasury]], axis=1).dropna()
            correlation, pvalue = stats.spearmanr(v.iloc[:,0], v.iloc[:,1])
            if pvalue<0.01:
                res= res.append(pd.Series({'fr007':each_fr007, 'treasury':each_treasury,"lead_lag":lead_lag,"correlation": correlation, 'data_length':len(v)}),
                          ignore_index=True)
res
```
